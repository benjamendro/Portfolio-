```python
#for many cases we would like to know the most intence rain event in our area of interset
#out data located in the IMS data set
#the follwing code will help us to sort the most intence rain events from the IMS data
#we should provide the code a CSV file that can be downloaded easy from the IMS site
#the CSV should have column of station name, date and rainfall amount

#importing the data base
import pandas as pd
data = pd.read_csv(r"d:\Users\97252\Downloads\data_202402281936.csv")
data.head()

#to avoid errors we will change the column name to English

new_names={"תחנה" : "station","תאריך": "date", 'כמות הגשם היומית (מ"מ)': "rain"}
data = data.rename(columns=new_names)

#change the date column to datetime format

data["date"] = pd.to_datetime(data["date"], format='%d/%m/%Y')
data.head()

#grouping the data by the station and 

grouped = data.groupby("station")
station_dataframes = {}
for station_name,station_df in grouped:
    station_dataframes[station_name] = station_df.copy()
    

#now we will create accumulated data frame with top 30 rain events for each station

top_30_events = []

# Iterate over each station
for station_name, station_df in station_dataframes.items():
    # Sort the DataFrame by date
    station_df.sort_values(by='date', inplace=True)
    
    # Calculate the rolling sum for a 3-day window
    rolling_sum = station_df.set_index('date').rolling(window=pd.Timedelta(days=3), min_periods=1).sum()
    
    # Add the rolling sum as a new column
    station_df['rolling_sum'] = rolling_sum.values
    
    # Get the top 30 events for this station
    top_30_station_events = station_df.sort_values("rolling_sum", ascending=False).head(30)
    
    # Append the top events for this station to the list
    top_30_events.append(top_30_station_events)

# Concatenate all top events into a single DataFrame
top_30_events_df = pd.concat(top_30_events)


```
